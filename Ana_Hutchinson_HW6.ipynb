## Question 2

```python
# Remove empty strings from the list of strings (5) list1 = [”Mike”, ””, ”Emma”,”Kelly”, ””, ”Brad”]

list1 = ["Mike", "", "Emma", "Kelly", "", "Brad"]
list1 = [s for s in list1 if s != ""]
print(list1)
```

## Question 3

```python
# Write a function that takes a list with duplicate elements and returns a new list with only unique elements.

def get_unique_elements(input_list):
    return list(set(input_list))

list_with_duplicates = [1, 2, 2, 3, 1, 4, 5, 4]
unique_list = get_unique_elements(list_with_duplicates)
print(unique_list)
```

## Question 4

```python
# Given a Python list, write a program to remove all occurrences of item 20.

list1 = [5, 20, 15, 20, 25, 50, 20]
list1 = [x for x in list1 if x != 20]
print(list1)
```

## Question 5

```python
# Use list comprehension to create a new list containing only the numbers from a given list.

my_list = [1, 2, 3, 'Jessa', 4, 5, 'Kelly', 'Jhon', 6]
numbers_list = [x for x in my_list if isinstance(x, (int, float))]
print(numbers_list)
```

## Question 6

```python
# Write a function to flatten a list of lists into a single, non-nested list. (e.g., [[1, 2], [3, 4]] becomes [1, 2, 3, 4]).

def flatten_list(list_of_lists):
    return [item for sublist in list_of_lists for item in sublist]

list_of_lists = [[1, 2], [3, 4], [5, 6, 7]]
flattened_list = flatten_list(list_of_lists)
print(flattened_list)
```
